Código 1: Diseñe una función en Python que cumpla con las siguientes condiciones:
Permita recibir como entrada diferentes tipos de datos (números enteros, decimales o texto).
 Verifique que el número sea natural (entero y mayor o igual a cero).
 Recorra los números desde 0 hasta m:
- Para 0 y 1, imprima el número.
- Para los demás, determine si son primos o no.
- Si no es primo, muestre una factorización básica.
Ejecute la función con la entrada m = 5 y m = "a"

```python
def analizar_numero(m):

    if isinstance(m, int) and m >= 0:
        for n in range(0, m + 1):

            if n == 0 or n == 1:
                print("El número es", n)

            else:
                es_primo = True

                for x in range(2, n):
                    if n % x == 0:
                        print(n, "no es primo:", n, "=", x, "*", n // x)
                        es_primo = False
                        break

                if es_primo:
                    print(n, "es un número primo")

    else:
        print("El número no es natural")
```

```python
analizar_numero(5)
```

```python
analizar_numero("a")
```

```python
analizar_numero(-3)
```

```python
analizar_numero(3.8)
```

```python
analizar_numero(6)
```

```python
analizar_numero(20)
```

m >= 0 asegura que sea natural, ademas se recorre desde 0 hasta m para cada número hasta llegar a ese número m, para las demás funciones se verifica si tiene divisores es decir el número no es primo, claramente si tiene divisores el número es primo, además si el número no es primo muestra una factorización básica.

Además de ser negativo o décimal el programa tiene como salida "no es número natural".

Código 2: Cree una función cuya entrada sea un objeto tipo str, el cual cuente:
 Vocales mayúsculas y minúsculas,
 Consonantes mayúsculas y minúsculas,
números y
otros caracteres (como espacios, números o sı́mbolos)
La salida de la función es un diccionario con la cantidad total de vocales, consonantes y otros
caracteres encontrados en el texto.
Ejecute la función con una cadena ingresada por el usuario.

```python
def contar_caracteres(texto):
    vocales = "aeiouAEIOU"
    consonantes = "bcdfghjklmnñpqrstvwxyzBCDFGHJKLMNÑPQRSTVWXYZ"

    cont_vocales = 0
    cont_consonantes = 0
    cont_numeros = 0
    cont_otros = 0

    for caracter in texto:
        if caracter in vocales:
            cont_vocales += 1
        elif caracter in consonantes:
            cont_consonantes += 1
        elif caracter.isdigit():
            cont_numeros += 1
        else:
            cont_otros += 1

    return {
        "vocales": cont_vocales,
        "consonantes": cont_consonantes,
        "numeros": cont_numeros,
        "otros": cont_otros
    }


cadena = input("Ingrese una cadena de texto: ")
resultado = contar_caracteres(cadena)

print("Resultados:")
for clave, valor in resultado.items():
    print(clave, ":", valor)
```

inicialmente se definen vocales y consonantes en mayúsculas y minúsculas y se recorren todos los caracteres del texto.

además las siguientes funciones se usan de la siguiente forma,
 in para identificar letras, isdigit() para números.

si no se indican letras o caracteres se considera otros caracteres (espacios, símbolos, etc.), finalmente Se devuelve un diccionario con los conteos y es importante contemplar mayúsculas y minúsculas para evitar errores.
