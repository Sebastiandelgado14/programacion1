```python
%%writefile promedio.py
import statistics

def calcular_promedio(n1, n2, n3, n4):
    return (n1 + n2 + n3 + n4) / 4

def mediana(n1, n2, n3, n4):
    return statistics.median([n1, n2, n3, n4])

def moda(n1, n2, n3, n4):
    try:
        return statistics.mode([n1, n2, n3, n4])
    except:
        return "No hay moda única"

def estado(promedio):
    if promedio >= 3.0:
        return "Aprobado"
    else:
        return "No aprobado"
```

```python
import pedirnotas
import promedio


nombre = pedirnotas.pedir_nombre()
n1, n2, n3, n4 = pedirnotas.pedir_notas()

prom = promedio.calcular_promedio(n1, n2, n3, n4)
med = promedio.mediana(n1, n2, n3, n4)
mod = promedio.moda(n1, n2, n3, n4)
est = promedio.estado(prom)


print(" RESULTADOS ")
print("Estudiante:", nombre)
print("Notas:", n1, n2, n3, n4)
print("Promedio:", prom)
print("Mediana:", med)
print("Moda:", mod)
print("Estado:", est)
```
