```python
#(1) Cambie una lista a un una tupla, a un diccionario y a un conjunto.

lista = [1, 2, 3, 3, 6, 14, 99 ]

tupla_de_lista = tuple(lista)
diccionario_de_lista = dict(enumerate(lista))
conjunto_de_lista = set(lista)

print("Lista principal:", lista)
print("Tupla:", tupla_de_lista)
print("Diccionario:", diccionario_de_lista)
print("Conjunto:", conjunto_de_lista)

```

```python
# (2) Similar para una tupla.
tupla = (10, 20, 30, 40)

lista_de_tupla = list(tupla)
diccionario_de_tupla = dict(enumerate(tupla))
conjunto_de_tupla = set(tupla)

print("Tupla :", tupla)
print("Lista:", lista_de_tupla)
print("Diccionario:", diccionario_de_tupla)
print("Conjunto:", conjunto_de_tupla)
```

```python
#(3) Similar para un diccionario
diccionario = { "Nombre": "Sebas", "Edad": "20", "hobby": "Fútbol","Código": "2230128"}

lista_apartir_dic = list(diccionario.values())
tupla_apartir_dic = tuple(diccionario.values())
conjunto_apartir_dic = set(diccionario.values())

print("Diccionario de sebas:", diccionario)
print("Lista:", lista_apartir_dic)
print("Tupla:", tupla_apartir_dic)
print("Conjunto:", conjunto_apartir_dic)
```

```python
#(4) Similar para un conjunto
conjunto = {45, 47, 66, 90}

lista_de_conjunto = list(conjunto)
tupla_de_conjunto = tuple(conjunto)
diccionario_de_conjunto = dict(enumerate(conjunto))

print("Conjunto inicial:", conjunto)
print("Lista:", lista_de_conjunto)
print("Tupla:", tupla_de_conjunto)
print("Diccionario:", diccionario_de_conjunto)
```
